```python
df = pd.read_excel("Fotage_Project.xlsx")
```

```python
df['Revenue Growth (%)'] = df.groupby(['company'])['total revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['company'])['net income'].pct_change() * 100
```

```python
# Fill NA values that result from pct_change calculations with 0 or an appropriate value
df.fillna(0, inplace=True)

# Display the dataframe to verify the calculations
print(df)

# Optionally, you could summarize these findings for each company
summary = df.groupby('company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean'
}).reset_index()

print("\nYear-over-Year Average Growth Rates (%):")
print(summary)
```

```python
def financial_chatbot(query):
    query = query.lower()

    # Example: Net income for Apple in 2023
    for company in df['company'].unique():
        for year in df['year of 10-K'].unique():
            if company.lower() in query and str(year) in query:
                row = df[(df['company'] == company) & (df['year of 10-K'] == year)].iloc[0]

                if "net income" in query:
                    return f"{company}'s net income in {year} was {row['net income']:,}."
                elif "revenue" in query:
                    return f"{company}'s total revenue in {year} was {row['total revenue']:,}."
                elif "assets" in query:
                    return f"{company}'s total assets in {year} were {row['total assets']:,}."
                elif "cash flow" in query:
                    return f"{company}'s cash flow in {year} was {row['cash flow']:,}."

    return "Sorry, I couldn't find an answer to that. Try asking about net income, revenue, or assets for a company and year."
```

```python
print(financial_chatbot("What was Apple's net income in 2023?"))
print(financial_chatbot("Tell me Tesla’s revenue in 2024"))
print(financial_chatbot("How much was Microsoft’s cash flow in 2022?"))

```
